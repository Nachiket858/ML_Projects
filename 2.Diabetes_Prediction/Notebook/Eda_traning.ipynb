```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv(r"E:\Python\Myenv\Data_Science\ML_Projects\2.Diabetes_Prediction\Dataset\diabetes.csv")
```

```python
df["Glucose"]=df["Glucose"].replace(0,df['Glucose'].mean())
df["BloodPressure"]=df["BloodPressure"].replace(0,df['BloodPressure'].mean())
df["SkinThickness"]=df["SkinThickness"].replace(0,df['SkinThickness'].mean())
df["Insulin"]=df["Insulin"].replace(0,df['Insulin'].mean())
df["BMI"]=df["BMI"].replace(0,df['BMI'].mean())
```

```python
### Dividing the dataset
X =  df.drop(columns='Outcome')
y= df['Outcome']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
X_train=scalar.fit_transform(X_train)
X_test=scalar.transform(X_test)
```

```python
import pickle
pickle.dump(scalar,open(r'E:\Python\Myenv\Data_Science\ML_Projects\2.Diabetes_Prediction\Models\Scalar.pkl','wb'))
```

```python
from sklearn.linear_model import LogisticRegression
classifire = LogisticRegression()
```

```python
classifire.fit(X_train,y_train)
```

```python
y_pred = classifire.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print(accuracy_score(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

# Hypertuning of the model
